```python
# https://kanoki.org/python-plotting-latitude-and-longitude-from-csv-on-map-using-basemap-folium-geopandas-and-plotly
#[Python Plotting latitude and longitude from csv on map using basemap, folium, geopandas and plotly](Python Plotting latitude and longitude from csv on map using basemap, folium, geopandas and plotly)

# importing the libraries
%matplotlib inline
import pandas as pd
from mpl_toolkits.basemap import Basemap    # pip install basemap
import matplotlib.pyplot as plt
```

```python
# setting the filepath and loading csv
countries = 'world_country_and_usa_states_latitude_and_longitude_values.csv'
data = pd.read_csv(countries)
data
```

```python
# setting the size of the map
fig = plt.figure(figsize=(12,9))


# creating the map - setting latitude and longitude
m = Basemap(projection = 'mill', llcrnrlat = -90, urcrnrlat = 90, llcrnrlon = -180, urcrnrlon = 180, resolution = 'c')
```

```python
# drawing the coastline
m.drawcoastlines()
m.drawcountries(color='gray')
m.drawstates(color='gray')

# creating variable for latitude and longitude to list
lat = data['latitude'].tolist()
lon = data['longitude'].tolist()

# plotting the map
m.scatter(lon, lat, latlon = True, s = 10, c = 'red', marker = 'o', alpha = 1)

plt.show()
```
